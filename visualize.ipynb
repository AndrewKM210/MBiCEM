```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
log_pc0 = pickle.load(open("logs/halfcheetah_medium_expert_pc0.pkl", "rb"))
log_pc20 = pickle.load(open("logs/halfcheetah_medium_expert_pc20.pkl", "rb"))
plt.figure()
plt.plot(np.cumsum(log_pc0[0].get_dict()["rewards"]), label="MBiCEM no penalization")
plt.plot(np.cumsum(log_pc20[0].get_dict()["rewards"]), label="MBiCEM")
plt.legend()
plt.xlabel("Timestep")
plt.ylabel("Return")
plt.title("halfcheetah-medium-expert-v0")
plt.grid(alpha=0.5)
plt.show()
```
